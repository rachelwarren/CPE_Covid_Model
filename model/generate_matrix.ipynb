```python
import pandas as pd
import numpy as np
```

```python
#expanded matrix

ls = []
for race in ['White', 'Black']:
    for labour_type in ['Forced_Labour', 'Non_Essential_Labour', 'Police']:
        for loc in ['At_Work', 'At_Home']:
            simple_name = f'{race}_{labour_type}' if labour_type != 'Non_Essential_Labour' else race
            
            ls.append(
                {'Race' : race,
                 'Labour_Type' : labour_type,
                 'Location': loc,
                'Sub_Group_Name': f'{race}_{labour_type}_{loc}',
                'Group_Name': simple_name})
    
    ls.append({'Race': race, 'Labour_Type': 'Non_Working', 'Location': 'At_Home',
                     'Sub_Group_Name': f'{race}_Non_Working', 'Group_Name': race})
    ls.append({'Race': race, 'Labour_Type': 'Prison',
                     'Sub_Group_Name': f'{race}_Prison', 'Group_Name': f'{race}_Prison'})

group_df = pd.DataFrame(ls)
group_df = group_df.set_index('Sub_Group_Name')
group_df
```

```python
# Share of Population 

group_sizes = {} #dictionary with size of the final population groups that we use 
WHITE_PROP = .6040
BLACK_PROP = 0.1340


WHITE_FORCED_LABOUR_PROP = 0.1125  #proprotion of all white residents 
BLACK_FORCED_LABOUR_PROP = 0.1362 #Proportion of all black residents


#?????

POLICE_PROP_WHITE = 0.597  #proportion of police who are white
POLICE_PROP_BLACK = 0.215 #proportion of police who are black

WHITE_POLICE_PROP = 0.00167846 #B20 #proporiton of white who are police 
BLACK_POLICE_PROP = 0.00272 

group_sizes['White_Police'] = WHITE_POLICE_PROP * WHITE_PROP
group_sizes['Black_Police'] = BLACK_POLICE_PROP * BLACK_PROP

WHITE_NON_ESSENTIAL_LABOUR_PROP = 0.3749 # B23
BLACK_NON_ESSENTIAL_LABOUR_PROP = 0.30438 #B22 prop of black residents doing non essential labour


#????????????
#These are pull from the calculations spread sheet I don't know if they are write and I have not used the full calculations
WHITE_INCARCERATED_PROP = 0.00406 #Proportion of all white people incarcerated Calculations B 32
BLACK_INCARCERATED_PROP = 0.01632 #Proportion of all black people who are incarcerated Calculations B33

group_sizes['White_Prison'] = WHITE_INCARCERATED_PROP * WHITE_PROP
group_sizes['Black_Prison'] = BLACK_INCARCERATED_PROP * BLACK_PROP

white_non_working = 1 - WHITE_FORCED_LABOUR_PROP - WHITE_INCARCERATED_PROP - WHITE_NON_ESSENTIAL_LABOUR_PROP
black_non_working = 1 - BLACK_FORCED_LABOUR_PROP - BLACK_INCARCERATED_PROP - BLACK_NON_ESSENTIAL_LABOUR_PROP

group_sizes['White'] = WHITE_NON_ESSENTIAL_PROP * WHITE_PROP
group_sizes['Black'] = BLACK_NON_ESSENTIAL_PROP * BLACK_PROP

group_sizes['White_Forced_Labour'] = WHITE_PROP * (WHITE_FORCED_LABOUR_PROP)
group_sizes['Black_Forced_Labour'] = BLACK_PROP * (BLACK_FORCED_LABOUR_PROP)

group_df['Group_Size'] = group_df['Group_Name'].replace(group_sizes)

group_df['Proportion_of_Racial_Group'] = group_df.index
group_df['Proportion_of_Racial_Group'] = group_df['Proportion_of_Racial_Group'].apply(lambda x: {
    'White_Forced_Labour_At_Home' : WHITE_FORCED_LABOUR_PROP,
    'Black_Forced_Labour_At_Home' : BLACK_FORCED_LABOUR_PROP,
    'White_Non_Essential_Labour_At_Home' : WHITE_NON_ESSENTIAL_LABOUR_PROP,
    'Black_Non_Essential_Labour_At_Home' : BLACK_NON_ESSENTIAL_LABOUR_PROP,
    
    'White_Non_Working' : white_non_working,
    'Black_Non_Working' : black_non_working,
    
    'White_Police_At_Home': WHITE_POLICE_PROP,
    'Black_Police_At_Home': BLACK_POLICE_PROP,
    'White_Prison': WHITE_INCARCERATED_PROP,
    'Black_Prison': BLACK_INCARCERATED_PROP
}.get(x)).fillna(0.0)

group_df[['Group_Name', 'Proportion_of_Racial_Group', 'Group_Size']].drop_duplicates()
```

```python
# Fill in Hours in Neighborhood

group_df['Hours_In_Neighborhood'] = np.where(
    (group_df['Labour_Type'] == 'Non_Working'), 112,
        np.where((group_df['Labour_Type'] != 'Non_Working') & 
                 (group_df['Location'] == 'At_Home'), 72, 0))

group_df[['Hours_In_Neighborhood']]                                                                              
```

```python
group_df['People_Hours'] = group_df['Hours_In_Neighborhood'] * group_df['Proportion_of_Racial_Group']
```
